```python
#INSTALL PLOTLY (PYTHON LIBRARY WITH HOVER FUNCTION)
!pip install plotly
import plotly.express as px
```

```python
#INSTALL PANDAS (DATA VISUALIZATION LIBRARY)
import pandas as pd
```

```python
#IMPORT DATA AS CSV
#MUST ADD TO FILES BEFORE RUNNING
df = pd.read_csv('SAS 9.13.csv')
```

```python
df['Info'] = '<br>' + df['Format'] + '<br>' + df['Topic'] + '<br>' + df['GFX']
```

```python
df.rename(columns={'Livestream position (minutes)': 'Time (min)'}, inplace=True)
```

```python
df.rename(columns={'Concurrent viewers ∆': 'Viewers ∆'}, inplace=True)
```

```python
df.head()
```

```python
# CREATE BAR CHART WITH EPISODE DATA
# EDIT CHART TITLE HERE
chart = px.bar(df, x='Time (min)', y='Viewers ∆', text='Info',
             title='SAS 9.13 Viewer Insights')

# Customize the hover information
chart.update_traces(texttemplate='%{text}', textposition='none')

# Update layout for better hover display
chart.update_layout(hovermode='x')

# SET COLOR
chart.update_traces(marker_color=['red' if val < 0 else 'blue' for val in df['Viewers ∆']])

# Show the plot
chart.show()









```
